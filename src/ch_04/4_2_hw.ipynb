1. 수백만 개의 특성을 가진 훈련 세트에서는 어떤 선형 회귀 알고리즘을 사용할 수 있을까요?
- 확률적 경사 하강법, 미니배치 경사 하강법

2. 훈련 세트에 있는 특성들이 각기 아주 다른 스케일을 가지고 있습니다. 이런 데이터에 잘 작동하지 않는 알고리즘은 무엇일까요? 그 이유는 무엇일까요? 이 문제를 어떻게 해결할 수 있을까요?
- 규제가 있는 선형 모델, 규제는 가중치를 제한하는 목적인데 스케일이 매우 큰 특성의 가중치를 규제를 통해 제한할 수 없으므로 지역 최적점에 수렴할 가능성이 있음, 이러한 문제를 데이터의 스케일을 조정하는 방법을 통해 극복

3. 경사 하강법으로 로지스틱 회귀 모델을 훈련시킬 때 지역 최솟값에 갇힐 가능성이 있을까요?
- 로지스틱 회귀 모델의 경우 비용함수는 볼록 함수로 최솟값이 1개이므로 갇힐 가능성 없음

4. 충분히 오랫동안 실행하면 모든 경사 하강법 알고리즘이 같은 모델을 만들어낼까요?
- 몇가지 하이퍼파라미터를 적절히 선택하면 같은 모델을 만들어낼 가능성 있음

5. 배치 경사 하강법을 사용하고 에포크마다 검증 오차를 그래프로 나타내봤습니다. 검증 오차가 일정하게 상승되고 있다면 어떤 일이 일어나고 있는 걸까요? 이 문제를 어떻게 해결할 수 있나요?
- 큰 학습률, 알고리즘의 발산 등의 문제가 있을 수 있음, 훈련 에러도 동시에 올라갈 경우 학습률의 문제 가능성 다분, 훈련 에러가 올라가지 않을 경우는 과대적합 가능성이 있으므로 훈련을 정지하는 것이 좋음

6. 검증 오차가 상승하면 미니배치 경사 하강법을 즉시 중단하는 것이 좋은 방법일까요?
- 해당 배치의 문제일 가능성이 있으므로 오랫동안 진전이 없을 때 중단하는 것이 유리

7. (언급한 것 중에서) 어떤 경사 하강법 알고리즘이 가장 빠르게 최적 솔루션의 주변에 도달할까요? 실제로 수렴하는 것은 어떤 것일까요? 다른 방법들도 수렴하게 만들 수 있나요?
- 확률적 경사 하강법은 한 번의 하나의 훈련 샘플을 사용하기 때문에 가장 빠름, 다음으로 미니배치 GD, 경사 하강법도 시간만 충분하다면 가능성 있음

8. 다항 회귀를 사용했을 때 학습 곡선을 보니 훈련 오차와 검증 오차 사이에 간격이 큽니다. 무슨 일이 생긴 걸까요? 이 문제를 해결하는 세 가지 방법은 무엇인가요?
- 과적합의 가능성이 높기 때문에, 다항 차수를 낮추거나 모델 규제, 훈련 세트 크기 증가

9. 릿지 회귀를 사용했을 때 훈련 오차와 검증 오차가 거의 비슷하고 둘 다 높았습니다. 이 모델에는 높은 편향이 문제인가요, 높은 분산이 문제인가요? 규제 하이퍼파라미터 $\alpha$ 를 즐가시켜야 할까요, 아니면 줄여야 할까요?
- 과소적합 문제, 높은 편항을 가진 모델이므로 규제 하이퍼파리미터 $\alpha$ 를 감소

10. 다음과 같이 사용해야 하는 이유는?
    * 평범한 선형 회귀(즉, 아무런 규제가 없는 모델) 대신 릿지 회귀
    - 일반적으로 규제가 없는 모델보다 성능이 좋음
    * 릿지 회귀 대신 라쏘 회귀
    - 몇 개의 특정 특성에 연관되어 있다면 라쏘 회귀를 사용하여 특성 선택을 하는 것이 좋음
    * 라쏘 회귀 대신 엘라스틱넷
    - 몇 개의 특성이 강하게 연관, 훈련 샘플보다 특성이 더 많을 때 엘라스틱넷이 라쏘보다 일반적으로 선호

11. 사진을 낮과 밤. 실내와 실외로 분류하려 합니다. 두 개의 로지스틱 회귀 분류기를 만들어야 할까요?
- 배타적인 클래스가 아니므로 두 개의 로지스틱 회귀 분류기 생성
